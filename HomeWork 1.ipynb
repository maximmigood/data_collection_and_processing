## Задание 2

**Работа будет состоять с недокументированным API. Нужно ввести релевантный запрос на сайте** (https://www.delivery-club.ru/search)<br>
(а) из предложенных точек с помощью API найти долю (в %) с бесплатной и платной доставкой. Для каждой категории рассчитать среднюю минимальную стоимость заказа. <br>
(б) для каждой из категорий из пункта (а) рассчитать долю (в %) магазинов и ресторанов

```python
import requests
from pprint import pprint
import json
import pandas as pd
import numpy as np
```

```python
srch_prod = 'Мороженное'
```

```python
url = 'https://api.delivery-club.ru/api1.2/vendors/search'
params = {'latitude': 55.7577374,
          'longitude': 37.6164793,
          'query': srch_prod
         }
```

```python
response = requests.get(url=url, params=params)
```

```python
if response.ok:
    data = response.json()
```

```python
data
```

```python
shops_and_cafe = data['vendors']
```

```python
pprint(shops_and_cafe[0])
```

```python
for shop in shops_and_cafe:
    print(f"{shop['categoryId']} {shop['name']} {shop['delivery']['price']['value']} {shop['delivery']['minOrderPrice']['value']}")
```

```python
dataset = [(shop['name'], 
            shop['categoryId'], 
            shop['delivery']['price']['value'], 
            shop['delivery']['minOrderPrice']['value']) 
           for shop in shops_and_cafe]
```

```python
df = pd.DataFrame(dataset, columns=('from', 'category', 'delivery_price', 'delivery_min_order'))
```

```python
df.head()
```

```python
df['free_delivery'] = df['delivery_price'] == 0
```

```python
df.head()
```

---------------------------

(а) из предложенных точек с помощью API найти долю (в %) с бесплатной и платной доставкой. 

```python
print(f"всего магазинов - {len(df)}")
print(f"доля магазинов с платной доставкой - {len(df[df['free_delivery'] == False]) / len(df):.0%}")
print(f"доля магазинов с бесплатной доставкой - {len(df[df['free_delivery'] == True]) / len(df):.0%}")
```

----------------------

```python
gr = df.groupby("free_delivery").agg({"delivery_min_order": "mean"}).astype(int)
```

```python
gr
```

```python
gr.loc[False, 'delivery_min_order']
```

---------------

Для каждой категории рассчитать среднюю минимальную стоимость заказа.

```python
print(f"Средyяя минимальная стоимость заказа для бесплатной доставки - {gr.loc[True, 'delivery_min_order']} рубля")
print(f"Средyяя минимальная стоимость заказа с платной доставкой - {gr.loc[False, 'delivery_min_order']} рубля")
```

(б) для каждой из категорий из пункта (а) рассчитать долю (в %) магазинов и ресторанов

```python
df['type'] = np.where(df['category'] == 1, 'Ресторан', 'Магазин')
```

```python
gb = df.pivot_table()
```

```python
pivot = pd.pivot_table(df, index='type', columns='free_delivery', values='from', aggfunc='count', margins=True)
```

```python
pivot
```

```python
print(f"Магазинов с платной доставкой - {pivot.loc['Магазин', False] / pivot.loc['Магазин', 'All']:.2%}")
print(f"Магазинов с бесплатной доставкой - {pivot.loc['Магазин', True] / pivot.loc['Магазин', 'All']:.2%}")
```

```python
print(f"Ресторанов с платной доставкой - {pivot.loc['Ресторан', False] / pivot.loc['Ресторан', 'All']:.2%}")
print(f"Ресторанов с бесплатной доставкой - {pivot.loc['Ресторан', True] / pivot.loc['Ресторан', 'All']:.2%}")
```
