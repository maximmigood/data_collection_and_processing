Структура данных в виде словаря должна содержать:
- название источника;
- наименование новости;
- ссылку на новость;
- дата публикации.

Парсинг сделан для dzen.ru

```python
from lxml import html
import requests
```

```python
url = 'https://dzen.ru/news/'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.5112.124'}
params = {'sso_failed': 'blocked'}
```

```python
r = requests.get(url=url, headers=headers, params=params)
```

```python
r.status_code
```

```python
r.url
```

```python
dom = html.fromstring(r.text)
```

```python
news = dom.xpath('//div[contains(@class,"mg-grid__item")]')
```

```python
news_list = []
for n in news:  #
    a = {}
    name = n.xpath('.//a[@class="mg-card__link"]/text()')
    if name:
        link = n.xpath('.//a[@class="mg-card__link"]/@href')[0]
        name = name[0]
        time = n.xpath('.//span[@class="mg-card-source__time"]/text()')[0]
        src = n.xpath('.//a[@class="mg-card__source-link"]/text()')[0]
        a['src'] = src
        a['name'] = name
        a['link'] = link
        a['time'] = time
        news_list.append(a)
    else:
        continue
```

```python
print(f'Количество новостей - {len(news_list)}')
for nws in news_list:
    print(nws['src'])
    print(nws['name'])
    print(nws['link'])
    print(nws['time'])
    print('-' * 80)
```
